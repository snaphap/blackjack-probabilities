## True Count and Expected Value ##

### Table generation

```python
def roundCount(count, precision = 1):
    return (1 / precision) * round(precision * count)

def resultToEV(result):
    if 'Lose' in result:
        output = -1
    elif 'Push' in result:
        output = 0
    elif 'Win' in result:
        output = 1
    elif 'Blackjack' in result:
        output = 3/2
    else:
        output = 0
        print(result)
        #raise Exception(f'that shit {result} is not a result')
    if 'Doubled' in result:
        output *= 2
    return output

def averageEV(results):
    return np.mean(np.array([resultToEV(result) for result in results]))
```

```python
gameDataRoundedTC = gameData.with_column('Rounded True Count', gameData.apply(roundCount, 'True Count'))
counts = gameDataRoundedTC.group('Rounded True Count').column('count')
gameDataEV = gameDataRoundedTC.select('Rounded True Count', 'Result').group('Rounded True Count', averageEV).with_column('Count', counts)
filteredGameDataEV = gameDataEV.where('Rounded True Count', are.between(-8, 8))
filteredGameDataEV
```

## Hand Type Stuff

### Initial stuff

```python
import ast

def handListToType(handList):
    handList = ast.literal_eval(handList)
    if handList == ['A', 'A']:
        return 'A,1'
    elif 'A' in handList:
        notAce = [card for card in handList if card != 'A'][0]
        return (f'A,{notAce}')
    else:
        return sum([int(card) for card in handList])
    
handListToType("['10', 'A']")
```

```python
handTypes = gameDataRoundedTC.apply(handListToType, 'Player Hand')
gameDataHandTypes = gameDataRoundedTC.with_column('Hand Type', handTypes)
gameDataHandTypes
```

### Data given true count

```python
def TCHT(trueCount, handType):
    """Returns a string that contains both the true count and hand type for a row."""
    return f'{handType}|{trueCount}'

def TCHTtoHT(TCHT):
    "Given the HT|TC format, returns hand type."
    pipeIndex = TCHT.index('|')
    return TCHT[:pipeIndex]

def TCHTtoTC(TCHT):
    "Given the HT|TC format, returns true count."
    pipeIndex = TCHT.index('|')
    return float(TCHT[(pipeIndex + 1):])

def hardOrSoft(handType):
    try:
        int(handType)
        return 'Hard'
    except:
        return 'Soft'
```

```python
gameDataTCHT = gameDataHandTypes.with_column('Hand Type|True Count', gameDataHandTypes.apply(TCHT, 'Rounded True Count', 'Hand Type'))
gameDataTCHT
```

```python
groupedGDTCHT = gameDataTCHT.select('Hand Type|True Count', 'Result').group('Hand Type|True Count', averageEV)
TCHTHandTypes = groupedGDTCHT.apply(TCHTtoHT, 'Hand Type|True Count')
TCHTTrueCounts = groupedGDTCHT.apply(TCHTtoTC, 'Hand Type|True Count')
groupedGDTCHT = groupedGDTCHT.with_columns(
    'Hand Type', TCHTHandTypes,
    'True Count', TCHTTrueCounts
)
groupedGDTCHT = groupedGDTCHT.with_column('Hard or Soft Total', groupedGDTCHT.apply(hardOrSoft, 'Hand Type'))
groupedGDTCHT
```

```python
boundary = 6
filteredGDTCHT = groupedGDTCHT.where('True Count', are.between(-1*boundary, boundary+1))
filteredGDTCHT
```

## Hard Totals

```python
filteredGDTCHThardTotals = filteredGDTCHT.where('Hard or Soft Total', are.equal_to('Hard'))
filteredGDTCHThardTotals.show(255)
```

```python
plt.style.use('bmh')

restrictedTrueCounts = [i for i in range(-1, -1*boundary-1, -1)] + [i for i in range(0, boundary+1)]
hardTotalColumns = Table().with_column('True Counts', restrictedTrueCounts)
for i in range(4, 21):
    averageEVs = filteredGDTCHThardTotals.where('Hand Type', are.equal_to(str(i))).column('Result averageEV')
    hardTotalColumns = hardTotalColumns.with_column(f'{i} Hard Total', averageEVs)
hardTotalColumns = hardTotalColumns.sort('True Counts')
    
fig, ax = plt.subplots()
colors = ['lightcoral', 'maroon', 'red', 'darkorange', 
                         'darkgoldenrod', 'gold', 'gray', 'olivedrab', 'chartreuse',
                   'darkgreen', 'turquoise', 'lightseagreen', 'steelblue',
                   'midnightblue', 'mediumpurple', 'violet', 'fuchsia']
np.random.shuffle(colors)
ax.set_prop_cycle(color=colors)

for column in hardTotalColumns.drop('True Counts'):
    line = ax.plot(range(-1*boundary, boundary+1), hardTotalColumns[column], label = column)
ax.legend(loc = 'center', bbox_to_anchor = (1.2, .5))
ax.set(xlabel = 'True Count', ylabel = 'Expected Value')

plt.show()
```

```python
plt.style.use('default')

data = []
for row in hardTotalColumns.drop('True Counts').rows:
    data += [list(row)]
data = np.array(data).transpose()
fig, ax = plt.subplots()
ax.set_xticks(np.arange(hardTotalColumns.num_rows), labels=[f'{count}' for count in hardTotalColumns.column('True Counts')])
ax.set_yticks(np.arange(hardTotalColumns.num_columns - 1), labels=hardTotalColumns.drop('True Counts').labels)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

im = ax.imshow(data, cmap = 'viridis', vmin = -.6, vmax = .6) #  viridis

for i in range(hardTotalColumns.num_columns - 1):
    for j in range(hardTotalColumns.num_rows):
        text = ax.text(j, i, round(data[i, j], 2),
                       ha="center", va="center", color="w", size = 6)

plt.xlabel('True Count')
plt.title('Expected earnings (bet size of 1)')
fig.tight_layout()
plt.show()
```

## Testing stuff

```python
import pandas as pd
import matplotlib.pyplot as plt
df = gameDataHandTypes.to_df()
df = df[['Upcard', 'Hand Type', 'Result']]
df['Result Vals'] = gameDataHandTypes.apply(resultToEV, 'Result')
df.drop(columns = ['Result'], inplace = True)
g = df.groupby(['Upcard', 'Hand Type'], group_keys = True).mean()

g
```

```python
plt.style.use('default')

upcardList = ['10', '2', '3', '4', '5', '6', '7', '8', '9', 'A']

data = {'10':[], '2':[], '3':[], '4':[], '5':[], '6':[], '7':[], '8':[], '9':[], 'A':[]}
handTypes = []
for index, row in g.iterrows():
    upcard = index[0]
    handType = index[1]
    if upcard == '10':
        handTypes.append(handType)
    data[upcard] += [row.iloc[0]]
 
dataDF = pd.DataFrame(data = data, index = handTypes)
# if this isnt here then pandas throws a warning and i still have no idea why
dataDF.loc[:, 'Index'] = 0

dataDF
```

```python
hardDataDF = dataDF.filter(regex = r'\A\d', axis=0)
hardDataDF.loc[:, 'Index'] = [int(i) for i in list(hardDataDF.index)]
hardDataDF.set_index('Index', inplace = True)
hardDataDF = hardDataDF.sort_index()
hardDataDF = hardDataDF.reindex(columns = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'A'])
hardDataDF
```

```python
fig, ax = plt.subplots()
im = ax.imshow(hardDataDF)

sortedUpcardList = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'A']

ax.set_xticks(np.arange(len(upcardList)), labels=sortedUpcardList)
ax.set_yticks(np.arange(len(hardDataDF['2'])), labels=hardDataDF.index)

for i, hardTotal in enumerate(hardDataDF.index):
    for j, upcard in enumerate(sortedUpcardList):
        text = ax.text(j, i, (result := round(hardDataDF.at[hardTotal, upcard], 2)),
                       ha="center", va="center", color="w" if result < .5 else 'black', fontsize = 6)

plt.xlabel('Upcard')
plt.ylabel('Hard total')
plt.title('Expected earnings given upcard and hard total')        

plt.show()
```

```python
softDataDF = dataDF.filter(regex = ',+', axis=0).drop('Index', axis = 1)

a = softDataDF.iloc[[i for i in range(len(softDataDF.index)) if i != 1], :]
b = softDataDF.iloc[[1], :]

softDataDF = pd.concat([a, b])
softDataDF = softDataDF.reindex(columns = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'A'])
softDataDF.drop('A,1', axis = 0, inplace = True)
softDataDF
```

```python
fig, ax = plt.subplots()
im = ax.imshow(softDataDF)

sortedUpcardList = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'A']

ax.set_xticks(np.arange(len(upcardList)), labels=sortedUpcardList)
ax.set_yticks(np.arange(len(softDataDF['2'])), labels=softDataDF.index)

for i, softTotal in enumerate(softDataDF.index):
    for j, upcard in enumerate(sortedUpcardList):
        text = ax.text(j, i, (result := round(softDataDF.at[softTotal, upcard], 2)),
                       ha="center", va="center", color="w" if result < .5 else 'black', fontsize = 6)

plt.xlabel('Upcard')
plt.ylabel('Soft total')
plt.title('Expected earnings given upcard and soft total')        

plt.show()
```
